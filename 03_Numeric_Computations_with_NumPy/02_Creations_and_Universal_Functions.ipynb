# Arithmetic Operations with NumPy Arrays

```python
list_1 = [1, 2, 3] # Python Built-in List
list_2 = [1, 2, 3]
```

```python
list_1 + list_2 # Addition Operation
```

Subtraction and other operations are not possible between Lists. Also adding a list with another list simply joins the list together.

Any arithmetic operations between equal-size arrays applies the operation element-wise also called Vectorization.

```python
arr_1 = np.arange(1,4)
arr_2 = np.arange(11,14)
print(arr_1)
print(arr_2)
```

```python
arr_1 + arr_2 # Addition Operation
```

```python
arr_1 - arr_2 # Subtraction Operation
```

```python
arr_1 * arr_2 # Multiplication Operation
```

```python
arr_1 / arr_2 # Division Operation
```

Arithmetic operations with scalars propagate the scalar argument to each element in the array:

```python
arr = np.array([[1., 2., 3.], [4., 5., 6.]])
```

```python
1 / arr
```

```python
arr / 10
```

```python
arr ** 0.5
```

```python
arr2 = np.array([[0., 4., 1.], [7., 2., 12.]])
```

```python
arr2 > arr # Element wise Operation
```

# View vs Copy

```python
arr = np.arange(10) # Creating a 1 NumPy Array
```

```python
arr
```

```python
arr_slice = arr[1:4] # Creating a Slice that is essentially a View
```

```python
arr_slice
```

```python
arr_slice[1] = 100 # Changing Element at Index 1 of the Slice
```

```python
arr # Reflected in Original Array
```

```python
arr_copy = arr[1:4].copy() # Creating a Copy of the Slice
```

```python
arr_copy[1] = 200 # Changing an Element from the Copy
```

```python
arr_copy
```

```python
arr
```

# Universal Functions

```python
arr = np.arange(10)
arr
```

```python
arr_2 = arr.reshape(2,5)
arr_2
```

```python
print(np.max(arr)) # Sum 1D Array
```

```python
print(np.sum(arr_2, axis=0)) # Collapse the Rows
```

```python
print(np.sum(arr_2, axis=1, keepdims=True)) # Collapse the Columns while keeping it 2D
```

```python
print(np.sum(arr_2)) # Sum 2D Array
```

```python
print(np.min(arr_2, axis=0)) # find the minimum along each Column i.e. Collapse all rows
```

```python
print(np.sqrt(arr_2))
```

# Random

```python
np.random.random(size=(2,3)) # Draw from Continous Uniform Distribution of 0 and 1
```

```python
np.random.randint(low=3, high=10, size=(2,3)) # Draw from Discrete Uniform between 3 and 10
```

```python
np.random.randn(2,3) # Draw from Standard Normal Mean of 0 and St.Dev 1
```

```python
np.random.normal(loc=5, scale=2, size=(2,3)) # Draw from Normal Distribution with mean of 5 and st. dev of 2
```
