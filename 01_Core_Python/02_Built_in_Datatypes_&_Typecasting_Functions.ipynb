# What are Built-in Data Types ?

A built-in datatype in Python is a data type that is pre-defined and readily available within the language itself, without needing to import any external libraries.

We are going to discuss enough to get ourselves ready for Python.

Our aim in this Notebook is only recognise a Built-in Data Type when we see it.

## Numeric Datatypes

```python
print(5) # Integer
print(5.0) # Float
```

## Boolean Datatype

```python
print(True) # Boolean
print(False) # Boolean
```

## Sequence Data Types

```python
print("Hello") # string
```

```python
print([1,2,3,'a','b','c']) # list # ordered # mutable
```

```python
print((10, 20, 30, 15.50)) # tuple # ordered # immutable
```

```python
print({1,1,2,3,5}) # Set # unordered # mutable
```

# Type Conversion Functions

In Python, a type conversion function is a built-in function that changes a value from one data type into another.

This process is also known as Type Casting.

These functions are named after the data type you want to convert to. You pass the original value to the function, and it returns a new value of the target type.

## Integer Conversion

```python
x = 5  # Integer datatype stored inside variable x
print(float(x))     # int → float
```

```python
x = 5  # Integer datatype stored inside variable x
print(str(x))       # int → str
```

```python
x = 5  # Integer datatype stored inside variable x
print(bool(x))      # int → bool
```

```python
x = 0  # Integer datatype stored inside variable x
print(bool(x))      # int → bool
```

Integer Data type can not be converted into List, Tuple, Dictionary or Set.

## Float Conversion

```python
x = 3.14  # Float datatype stored inside variable x
print(int(x))       # float → int (decimal part truncated)
```

```python
x = 3.14  # Float datatype stored inside variable x
print(str(x))       # float → str
```

```python
x = 3.14  # Float datatype stored inside variable x
print(bool(x))      # float → bool
```

```python
x = 0.0  # Float datatype stored inside variable x
print(bool(x))      # float → bool
```

Float Data type can not be converted into List, Tuple, Dictionary or Set.

## String Conversion

```python
x = "123"  # String datatype stored inside variable x
print(int(x))       # str → int (only works if string is numeric)
print(float(x))     # str → float (only works if string is numeric)
```

```python
x = "Hello"  # String datatype stored inside variable x
print(bool(x))      # str → bool (False only if string is empty)
```

```python
x = ""  # String datatype stored inside variable x
print(bool(x))      # str → bool (False only if string is empty)
```

```python
x = "Python"  # String datatype stored inside variable x
print(list(x))       # str → list of characters
print(tuple(x))      # str → tuple of characters
print(set(x))        # str → set of unique characters
```

String Datatype can not be converted to Dictionary.

## Boolean Conversion

```python
x = True  # Boolean datatype stored inside variable x
print(int(x))       # bool → int (True → 1, False → 0)
print(float(x))     # bool → float (True → 1.0, False → 0.0)
```

```python
x = False  # Boolean datatype stored inside variable x
print(int(x))       # bool → int (True → 1, False → 0)
print(float(x))     # bool → float (True → 1.0, False → 0.0)
```

```python
x = True  # Boolean datatype stored inside variable x
print(str(x))       # bool → str ('True' or 'False')
```

```python
x = False  # Boolean datatype stored inside variable x
print(str(x))       # bool → str ('True' or 'False')
```

Boolean can not be converted into List, Tuple, Set or Dictionary.

## List Conversion

```python
x = [1, 2, 3]  # List datatype stored inside variable x
print(str(x))       # list → str
print(bool(x))      # list → bool (False only if list is empty)
```

```python
x = [1, 2, 3]  # List datatype stored inside variable x
print(tuple(x))     # list → tuple
print(set(x))       # list → set
```

```python
x = [['a', 1], ['b', 2]]  # List containing key-value pairs
print(dict(x))     # list → dict (works only if elements are key-value pairs)
```

List Datatype can not be converted into Integer or Float.

## Tuple Conversion

```python
x = (1, 2, 3)  # Tuple datatype stored inside variable x
print(list(x))     # tuple → list
print(set(x))      # tuple → set
```

```python
x = (1, 2, 3)  # Tuple datatype stored inside variable x
print(str(x))       # tuple → str
print(bool(x))      # tuple → bool (False only if tuple is empty)
```

```python
x = (('a',1), ('b',2))  # Tuple containing key-value pairs
print(dict(x))     # tuple → dict (works only if elements are key-value pairs)
```

Tuple can not be converted into Integer or Float.

## Set Conversion

```python
x = {1, 2, 3}  # Set datatype stored inside variable x
print(list(x))     # set → list
print(tuple(x))    # set → tuple
```

```python
x = {1, 2, 3}  # Set datatype stored inside variable x
print(str(x))       # set → str
print(bool(x))      # set → bool (False only if set is empty)
```

Set can not be converted into integer, float or dictionary.

## Dictionary Conversion

```python
x = {'a': 1, 'b': 2}  # Dictionary datatype stored inside variable x
print(list(x))     # dict → list (keys only)
print(tuple(x))    # dict → tuple (keys only)
print(set(x))      # dict → set (keys only)
```

```python
x = {'a': 1, 'b': 2}  # Dictionary datatype stored inside variable x
print(str(x))       # dict → str
print(bool(x))      # dict → bool (False only if dict is empty)
```

Dictionary can not converted into Integer or Float.

# range() Function

In Python, range is a built-in data type that represents an immutable sequence of numbers, and range() is the function used to create that sequence.

The Syntax of range() Function is as follows:

range(start, stop, step)

The Resultant Data Type is called Range Data Type which is immutable and efficient and can be used to create List, Tuple and Sets.

```python
range_1 = range(10) # Start from 0 and Stop before 10
range_2 = range(1,10) # Start from 1 and Stop before 10
range_3 = range(1,10,3) # Start from 1 and Stop before 10 and Stepsize is 3
```

```python
list_1 = list(range_1)
tuple_2 = tuple(range_2)
set_3 = set(range_3)

print(list_1)
print(tuple_2)
print(set_3)
```

```python
range_1 = range(5)
print(type(range_1))
```

# len() Function

len() returns the number of items inside a collection.

```python
print(len('Haris'))
print(len('  '))
print(len(''))
```

```python
print(len(['Haris',1,2.5]))
```

```python
print(len((7.57,'Good')))
```

```python
print(len({1,2,3,4,5}))
```

```python
print(len({'name':'Haris','age':35})) # Counts total number of Keys
```

len() does not work with Integer or Float or Boolean.
