# Commas

There are situations where we might want to use print() Function multiple times.

Each print() instruction outputs the text in a new line.

```python
print(20+3)
print("Hi")
print(19.75)
```

A Line of Code is called "Statement".

When you give Computers many Statements, the Instruction will be executed line by line, from Top to Bottom.

The print() Functions can take in multiple values to output on the same line.

This can be done by seperating them with Commas.

```python
print("Balance:", 500)
```

```python
print("We need to give them",500,"US Dollars.")
```

By default, the print() function separates arguments with a space; however, this can be customized by passing an argument using the "sep=" parameter.

```python
print("Iron","Man",sep="-")
```

```python
print("companyname","email.com",sep="@")
```

# Variables

Variables act like a Box in which we can store Python Objects like Strings, Integers and Floats.

```python
age = 31 # Assigning Integer Object to Variable "age"
hourly_wage = 20.5 # Assigning Float Object to Variable "hourly_wage"
print("The Hourly Wage offered by this Company is",hourly_wage,"and minimum age of Qualification is",age)
```

```python
country = "Panama" # Assigning String Object to Variable "country"
print("Country Name:", country)

country = "Mexico" # Assigning a New String Object to Variable "country"
print("Country Name:", country)
```

```python
# We already have Integer Object 31 stored inside Variable "age"
print("You shall be",age + 2,"years old after 2 Years!")
```

```python
# adding two Numerics
print(age + hourly_wage)
```

By the way, "=" is called the Assignment Operator !

Assignment Operator is used to assign Python Objects to Variables!

```python
x = 5 # Assigning Integer Object to Variable "x"

x = x + 3 # Adding 3 in Variable "x" and reassigning this new output Python Object to the same Variable x

print(x)

x = x * 2 # Multiplying 2 with Variable "x" and reassigning this new output Python Object to the same Variable x

print(x)
```

Variables names have certain conditions !
1. We may use Alphabets, Numbers and only 1 Special Character i.e., underscore
2. Variable Name can not begin with Numbers
3. Python is case sensitive therefore x and X are not the same Variables.
4. Python Keywords and Function should not be used as Variable Names.

# id() Function

id() Function actually returns the memory address where the object is currently stored in your computer's RAM (specifically, the virtual memory address used by the Python process).

The primary use of id() in everyday Python programming is to check if two variables refer to the exact same object in memory

```python
x = 1000
y = 1000

print("Memory Address of Python Object associated with Variable x:", id(x))
print("Memory Address of Python Object associated with Variable y:", id(y))
```

Although the content of Python Object associated with Variable x and y is the same. They are stored in different Memory Address.
